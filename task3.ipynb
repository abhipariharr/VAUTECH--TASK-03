# Task: Data Cleaning & Preprocessing

```python
## Import module
import pandas as pd
```

```python
## Importing csv file
df = pd.read_csv("suicide_data_cleaned.csv")
```

```python
df.head()
```

```python
## Handling Missing values
df.isnull()  ## already handeled missing values in task 2
```

```python
## checking duplicate values
df.duplicated().any()
```

```python
## Count Duplicate values
df.duplicated().sum()
```

```python
## Show duplicate values
df[df.duplicated()]
```

```python
## Removing Duplicate values
df = df.drop_duplicates()
```

```python
## Checking any duplicate values
df.duplicated().any()
```

```python
## Convert categorical data into usable formats

## One-Hot coding - converts categorical (text) data into numerical format for analysis
## Creating seperate columns for each category

df = pd.get_dummies(df, columns=['sex', 'age', 'generation'])

#Dummy (in data analysis) means a temporary numerical variable created from categorical data so computers or machine learning models can understand it.
# helps machine learning model because ml models only understand numerical value not text format
```

```python
df.head()
```

```python
## Saving clean data set
df.to_csv('clean_suicide_data.csv', index=False)
```
